```python
import pandas as pd
```

```python
new = pd.read_json("new.json")
old = pd.read_json("old.json")
```

```python
merged = old.merge(new, on="solar", how="outer", indicator=True)
merged
```

```python
merged[merged["_merge"] != "both"]
```

```python
cols = old.columns
merged["diff"] = ""
merged["same"] = True
for col in cols:
    if col == "solar": continue
    diff = merged[f"{col}_x"] != merged[f"{col}_y"]
    merged["diff"][diff] += f"{col};"
    merged["same"][diff] = False
merged
```

```python
merged[~merged["same"] ]
```
